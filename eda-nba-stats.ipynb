```python
#histogram of top scores
hist = df["PTS"]

plt.hist(hist, bins=np.arange(40), alpha=1)
plt.grid(True)
plt.title("Histogram Of The Scores Within The Season")
plt.xlabel('Points')
```

```python
#histogram of top scores
hist = df["AST"]

plt.hist(hist, bins=25, alpha=1)
plt.grid(True)
plt.title("Histogram Of Number Of Assists")
plt.xlabel('Assists')
```

```python
#histogram of blocks

hist=df["BLK"]

plt.hist(hist, bins=30, alpha=1)
plt.grid(True)
plt.title("Histogram Of Number Of Blocks")
plt.xlabel('Blocks')
```

```python
#box plot free throwsmade

plt.figure(figsize=(14, 4))
plt.subplot(1, 5, 4)
sns.boxplot(y=df['FTM'],data =df, color='green')
plt.title('Box Plot Of Free Throwsmade')
```

```python
name_of_player = df.loc[df['PTS'].idxmax(), 'PLAYER']
name_of_player

top_points = df.nlargest(10,'PTS')

plt.bar(top_points['PLAYER'], top_points['PTS'])
plt.bar(name_of_player, df.loc[df['PLAYER'] == name_of_player, 'PTS'])
plt.xlabel('Player Name')
plt.ylabel('Points')
plt.title('Top 10 Scorers ')
plt.xticks(rotation=45, ha="right") 
plt.show()
```

```python
#team performance stats for rebounds mean
team_performance = df.groupby('TEAM'). agg({'REB':['mean']}).reset_index()
plt.bar(team_performance['TEAM'], team_performance['REB']['mean'], width=0.5)
plt.xticks(rotation=45, ha="right") 

plt.xlabel('Team Name')
plt.ylabel('No Of Rebounds')
plt.title('Teams And Their Rebounds')
plt.show()
```

```python
#leading teams and their number of 3 points made
team_performance = df.groupby('TEAM'). agg({'3PM':['mean']}).reset_index()
plt.bar(team_performance['TEAM'], team_performance['3PM']['mean'], width=0.5)
plt.xticks(rotation=45, ha="right") 

plt.xlabel('Team Name')
plt.ylabel('No Of 3 Points')
plt.title('Teams And Average 3 Points Made')
plt.show()
```

```python
#correlation between key performace 
df2=df[['REB','AST','STL','BLK', 'TOV']]

correlation = df2.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Of Performance Metrics')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FTM', y='FTA', data=df, alpha=0.5)
plt.title('FTM vs. FTA')
plt.xlabel('FTM')
plt.ylabel('FTA')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DREB', y='OREB', data=df, alpha=0.5)
plt.title('DREB vs. OREB')
plt.xlabel('OREB')
plt.ylabel('DREB')
plt.show()
```

```python
# Countplot for categorical variables how many players are on the team
sns.countplot(x='TEAM', data=df)
plt.xticks(rotation=90)
plt.title('Count Of Players By Team')
plt.show()
```

```python
sns.pairplot(df[['PTS','FGM','FGA','3PM','3PA']])
plt.show()
```

```python
#season and points
team_performance = df.groupby('SEASON'). agg({'PTS':['sum']}).reset_index()
plt.bar(team_performance['SEASON'], team_performance['PTS']['sum'], width=0.5)
plt.xticks(rotation=45, ha="right") 

plt.xlabel('Season')
plt.ylabel('Points Made')
plt.title('Points By Year')
plt.show()
```
